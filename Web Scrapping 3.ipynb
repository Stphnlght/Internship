### 2. Scrape Amazon.in

```python
product_url = []

for i in range(3):
    product_url_tags = driver.find_elements(By.XPATH,'//*[@class="a-link-normal s-no-outline"]')
    for i in product_url_tags:
        prod_url = i.get_attribute("href")
        product_url.append(prod_url)
    next_ = driver.find_element(By.XPATH, '//*[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_.click()
    time.sleep(5)
```

```python
product_name = []
for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        product_name_tags = driver.find_elements(By.XPATH,'//*[@class="a-size-large product-title-word-break"]')
        for i in product_name_tags:
            prod_name = i.text
            product_name.append(prod_name)
    except NoSuchElementException:
        product_name.append('-')
```

```python
brand_name = []
for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        bran_tags = driver.find_elements(By.XPATH,'//*[@class="a-size-large product-title-word-break"]')
        for i in bran_tags:
            brand_ = i.text.split(' ')[0:2]
            brand_name.append(brand_)
    except NoSuchElementException:
        brand_name.append('-')
```

```python
price = []
for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        price_ = driver.find_element(By.XPATH,'//*[@class="a-price-whole"]')
        price.append(price_.text)
    except NoSuchElementException:
        price.append('-')
```

```python
expect_delivery = []
for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        delivery = driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]')
        expect_delivery.append(delivery.text)
    except NoSuchElementException:
        expect_delivery.append('-')
```

```python
return_exchange = []
for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        exchange = driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]')
        return_exchange.append(exchange.text)
    except NoSuchElementException:
        return_exchange.append('-')
```

```python
availability = []
for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        available = driver.find_element(By.XPATH,'//*[@id="availability"]')
        availability.append(available.text)
    except NoSuchElementException:
        availability.append('-')
```

```python
df = pd.DataFrame({'Brand Name': brand_name, 'Product Name':product_name, 'Price': price, 'Availability': availability, 'Delivery': expect_delivery, 'Return/Exchange':return_exchange, 'Product Url':product_url})
df.head(10)
```

```python
df.to_csv('amazon_search.csv')
```

### 8. Billionares

```python
driver = webdriver.Chrome(r"C:/Users/ACER/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
driver.get('https://www.forbes.com/')
```

```python
action = ActionChains(driver)
side_bar = driver.find_element(By.XPATH, '//*[@class="hamburger_svg__fs-icon hamburger_svg__fs-icon--hamburger"]')
action.move_to_element(side_bar).click().perform()
time.sleep(2)
billionares = driver.find_element(By.XPATH, '//*[@class="B1S88Gxl"]')
action.move_to_element(billionares).click().perform()
time.sleep(2)
world = driver.find_element(By.XPATH, '/html/body/div[1]/header/nav/div[1]/div/div/div[2]/ul/li[1]/div[2]/div[3]/ul/li[1]/a')
action.move_to_element(world).click().perform()
```

```python
rank = []
name = []
net_worth = []
age = []
source = []
citizenship = []
industry = []

while True:
    ranks = driver.find_elements(By.XPATH,'//*[@class="rank"]')
    for i in ranks:
        rank.append(i.text)
    names = driver.find_elements(By.XPATH,'//*[@class="personName"]')
    for i in names:
        name.append(i.text)
    net_worths = driver.find_elements(By.XPATH,'//*[@class="netWorth"]')
    for i in net_worths:
        net_worth.append(i.text)
    ages = driver.find_elements(By.XPATH,'//*[@class="age"]')
    for i in ages:
        age.append(i.text)
    sources = driver.find_elements(By.XPATH,'//*[@class="source-text"]')
    for i in sources:
        source.append(i.text)
    country = driver.find_elements(By.XPATH,'//*[@class="countryOfCitizenship"]')
    for i in country:
        citizenship.append(i.text)
    industries = driver.find_elements(By.XPATH,'//*[@class="category"]')
    for i in industries:
        industry.append(i.text)
    try:
        next_ = driver.find_element(By.XPATH, '//*[@class="pagination-btn pagination-btn--next "]')
        next_.click()
        time.sleep(5)
    except:
        break
```

```python
df = pd.DataFrame({'Rank': rank, 'Name':name, 'Net Worth': net_worth, 'Age': age,
                   'Source': source, 'Citizenship':citizenship,'Industry': industry})
df.head(10)
```
